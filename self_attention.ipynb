```python
import torch
from torch.nn import functional as F
```

#### self attention as matrix multiplication??

attention as soft dictionary lookup and may be a differentiable lookup operation

```python
# b: batch size
# t: input size
# k: dimension size
b, t, k = 5, 5, 5
batch_sents = torch.randn(b, t, k)
batch_sents
```

```python
transpose_batch_sents = batch_sents.transpose(dim0=1, dim1=2)
transpose_batch_sents
```

```python
alpha = torch.matmul(batch_sents, transpose_batch_sents)
alpha
```

```python
att_scores = F.softmax(alpha, dim=2)
att_scores
```

```python
# sanity check
print(att_scores[0][0][0].item())
print((alpha[0][0][0].exp() / torch.sum(torch.tensor([t.exp() for t in alpha[0][0]]))).item())
```

```python
z = torch.matmul(att_scores, batch_sents)
z
```
